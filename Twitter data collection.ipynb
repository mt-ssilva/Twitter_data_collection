## TwitterSearch 
Mecanismo de pesquisa interno do Twitter

```python
from TwitterSearch import*

try:
    ts = TwitterSearch(#Object TwitterSearch
    consumer_key = consumer_key,
    consumer_secret = consumer_secret,
    access_token = access_token,
    access_token_secret = access_token_secret)
    
    tso = TwitterSearchOrder()#Object TwitterSearchOrder
    tso.set_keywords(['suic√≠dio','eu me sinto relaxado','me sinto bem','quero morrer','estou estressado','me sinto triste', 'mato-me','quero cometer suic√≠dio'], or_operator = True)
    tso.set_language('pt')
    
    for tweet in ts.search_tweets_iterable(tso): #ts.search_tweets_iterable(tso) trata-se de um metadata
        
        print('created_at: ',tweet['created_at'], 'User_id: ',tweet['id_str'], 'Tweet: ',tweet['text'])
        
        created_at = tweet['created_at']
        user_id = tweet['id_str']
        texto = tweet['text']
        
        with open("tweet.json", "a+")  as output:
            
            data = {"created_at": created_at,
                   "user_id": user_id,
                   "tweet": texto}
            
            #print(data)
            output.write("{}\n".format(json.dumps(data)))
            
except TwitterSearchException as e:
    print(e)
```

## NLTK - Natural Language Toolkit
Falando especificamente sobre processamento de linguagem natural, h√° uma ferramenta com diversos recursos voltados especificamente para isso, que √© a biblioteca NLTK (Natural Language Toolkit, que em portugu√™s significa Kit de Ferramentas de Linguagem Natural)

```python
import nltk
```

```python
def remove_stopWords(instancia):
    stopwords = set(nltk.corpus.stopwords.words('portuguese'))
    palavras = [i for i in instancia.split() if not i in stopwords]
    return(" ".join(palavras))
```

```python
from bs4 import BeautifulSoup
from nltk.tokenize import TweetTokenizer
import re

def cleaning_tweets(tweet):
    
    tweet = BeautifulSoup(tweet, "html.parser").get_text()
    tweet = re.sub(r"^a-zA-Z√†-√∫√Ä-√ö0-9", "", tweet.lower())
    tweet_tokenizer = TweetTokenizer()
    words = tweet_tokenizer.tokenize(tweet)
    
    return tweet
```

## Contagem de palavras

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df.cleand_tweets)
```

```python
#Cria o dataframe
word_count = pd.DataFrame(cv.get_feature_names(), columns = ['word'])
#Soma as palavras e converte para uma lista
word_count["count"] = count_matrix.sum(axis = 0).tolist()[0]
word_count = word_count.sort_values("count", ascending = False).reset_index(drop =True)
#AS 50 palavras mais usadas nos tweets
word_count[:50]
```

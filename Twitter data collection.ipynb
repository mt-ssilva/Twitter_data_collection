## NLTK - Natural Language Toolkit
Falando especificamente sobre processamento de linguagem natural, h√° uma ferramenta com diversos recursos voltados especificamente para isso, que √© a biblioteca NLTK (Natural Language Toolkit, que em portugu√™s significa Kit de Ferramentas de Linguagem Natural)

```python
def remove_stopWords(instancia):
    stopwords = set(nltk.corpus.stopwords.words('portuguese'))
    palavras = [i for i in instancia.split() if not i in stopwords]
    return(" ".join(palavras))
```

```python
from bs4 import BeautifulSoup
from nltk.tokenize import TweetTokenizer
import re

def cleaning_tweets(tweet):
    
    tweet = BeautifulSoup(tweet, "html.parser").get_text()
    tweet = re.sub(r"^a-zA-Z√†-√∫√Ä-√ö0-9", "", tweet.lower())
    tweet_tokenizer = TweetTokenizer()
    words = tweet_tokenizer.tokenize(tweet)
    
    return tweet
```
